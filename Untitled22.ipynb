```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

docs = [
    "Python AI Machine Learning Resume",
    "Software Engineer Resume",
    "Invoice Amount GST Payment",
    "Invoice Bill Payment",
    "Project Final Report",
    "Research Report Analysis"
]

labels = [
    "Resume",
    "Resume",
    "Invoice",
    "Invoice",
    "Report",
    "Report"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)

model = MultinomialNB()
model.fit(X, labels)

test = vectorizer.transform(
    ["Machine Learning Engineer Skills"]
)

print(model.predict(test))
```

```python
import os

os.makedirs("dataset/Resume", exist_ok=True)
os.makedirs("dataset/Invoice", exist_ok=True)
os.makedirs("dataset/Report", exist_ok=True)

print("Folders created successfully!")
```
